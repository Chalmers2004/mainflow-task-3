```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
```

```python
data=pd.read_csv('householdtask3.csv')
```

```python
data.head(10)
```

```python
#linechart

plt.plot(data['income'])
plt.plot(data['expenditure'])

#title
plt.title("line chart")

#xlabel and ylabel
plt.xlabel('income')
plt.ylabel('expenditure')

#legend
plt.legend(["income","expenditure"])

#result
plt.show()
```

```python
#barchart

plt.bar(data['own_prop'],data['own_wm_prop'])

#title
plt.title("BAR CHART")

#xlabel and y label
plt.xlabel('own_prop')
plt.ylabel('own_um_prop')

#result
plt.show()
```
